```python
import mglearn
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
mglearn.plots.plot_linear_regression_wave()
```

```python
from sklearn.linear_model import LinearRegression
X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LinearRegression().fit(X_train, y_train)
```

```python
print("lr.coef_:", lr.coef_)
print("lr.intercept_:", lr.intercept_)
```

```python
print("Training Set Point: {:.2f}".format(lr.score(X_train, y_train)))
print("Test Set Point: {:.2f}".format(lr.score(X_test, y_test)))
```

```python
X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)
lr = LinearRegression().fit(X_train,y_train)
```

```python
print("Training Set Point: {:.2f}".format(lr.score(X_train, y_train)))
print("Test Set Point: {:.2f}".format(lr.score(X_test, y_test)))
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train, y_train)
print("Training Set Point: {:.2f}".format(ridge.score(X_train,y_train)))
print("Test Set Point: {:.2f}".format(ridge.score(X_test,y_test)))
```

```python
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print("Training Set Point: {:.2f}".format(ridge10.score(X_train,y_train)))
print("Test Set Point: {:.2f}".format(ridge10.score(X_test,y_test)))
```

```python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print("Training Set Point: {:.2f}".format(ridge01.score(X_train,y_train)))
print("Test Set Point: {:.2f}".format(ridge01.score(X_test,y_test)))
```

```python
plt.plot(ridge10.coef_,'^', label="Ridge alpha=10")
plt.plot(ridge.coef_,'s', label="Ridge alpha=1")
plt.plot(ridge01.coef_,'v', label="Ridge alpha=0.1")

plt.plot(lr.coef_,'o', label="LinearRegression")
plt.xlabel("Coefficient List")
plt.ylabel("Coefficient Magnitude")
xlims = plt.xlim()
plt.hlines(0,xlims[0], xlims[1])
plt.xlim(xlims)
plt.ylim(-25,25)
plt.legend()
```
